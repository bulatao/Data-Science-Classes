```python
#Your code here ...
# create 3 dataframes with text data
import pandas as pd

df1 = pd.read_csv('datatest.txt')
df2 = pd.read_csv('datatest2.txt')
df3 = pd.read_csv('datatraining.txt')
```

```python
# use df3 to train a model
# create an X and Y
X = df3.drop(['date','Occupancy'], axis=1)
X.columns.values
y = df3.drop(['date', 'Temperature', 'Humidity', 'Light', 'CO2', 'HumidityRatio'], axis=1)
```

```python
# train a model or models
# Logistic Regression
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X,y)
print(model)
# make predictions
expected = y
predicted = model.predict(X)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
cm = metrics.confusion_matrix(expected, predicted)
# print(model.intercept_)
# print(model.coef_)
```

```python
## Naive Bayes Classification
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X,y)
print(model)
# make predictions
expected = y
predicted = model.predict(X)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
cm = metrics.confusion_matrix(expected, predicted)
```

```python
# Knn- classification
# k-Nearest Neighbor Classification
from sklearn.neighbors import KNeighborsClassifier
model.fit(X,y)
print(model)
# make predictions
expected = y
predicted = model.predict(X)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
cm = metrics.confusion_matrix(expected, predicted)
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X,y)
print(model)
# make predictions
expected = y
predicted = model.predict(X)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
cm = metrics.confusion_matrix(expected, predicted)
```

## HW2) Take Wine Quality dataset from data exploration to classifier

```python
"""
https://archive.ics.uci.edu/ml/datasets/Wine+Quality
"""
```

```python
# Your code here ...
# put files into dataframes
dfWhite = pd.read_csv('winequality-white.csv', sep=';') # add sep option since it uses semicolon
dfRed = pd.read_csv('winequality-red.csv', sep=';')
```

```python
dfWhite.head() # 
```

```python
# we can add a column called red
dfWhite['White'] = pd.Series(1, index=dfWhite.index)
dfRed['White']= pd.Series(0, index=dfWhite.index)
# make value 1 as yes and 0 is no
```

```python
dfRed.tail()
```

```python
dfWine = pd.concat([dfWhite, dfRed], ignore_index=False) # concat 
```

```python
dfWine.tail()
```

```python
dfWine.describe()
```

```python
# split data to X and y
# get columns
# create two dataframe copies
# drop unnecessary columns
dfWhite.columns.values
X = dfWine.drop(['White'], axis=1)
y = dfWine.drop(['fixed acidity', 'volatile acidity', 'citric acid','residual sugar', 'chlorides', 'free sulfur dioxide','total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol','quality'], axis=1)
```

```python
# train a model with X and y
# train a model or models
# Logistic Regression
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X,y)
print(model)
# make predictions
expected = y
predicted = model.predict(X)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
cm = metrics.confusion_matrix(expected, predicted)
# print(model.intercept_)
# print(model.coef_)
```

```python
## Naive Bayes Classification
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X,y)
print(model)
# make predictions
expected = y
predicted = model.predict(X)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
cm = metrics.confusion_matrix(expected, predicted)
```

```python
# Knn- classification
# k-Nearest Neighbor Classification
from sklearn.neighbors import KNeighborsClassifier
model.fit(X,y)
print(model)
# make predictions
expected = y
predicted = model.predict(X)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
cm = metrics.confusion_matrix(expected, predicted)
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X,y)
print(model)
# make predictions
expected = y
predicted = model.predict(X)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
cm = metrics.confusion_matrix(expected, predicted)
```
